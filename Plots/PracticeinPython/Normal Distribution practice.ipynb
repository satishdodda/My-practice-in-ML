```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
data = pd.read_csv("PythonFiles/wc-at.csv")
```

```python
def UVA_numeric(data):
    var_group = data.columns
    size = len(var_group)
    plt.figure(figsize = (7*size,3), dpi = 400)

    
    #looping for each variable
    for j,i in enumerate(var_group):

        # calculating descriptives of column
        mini = data[i].min()
        maxi= data[i].max()
        ran = data[i].max()-data[i].min()
        mean = data[i].mean()
        median = data[i].median()
        st_dev = data[i].std()
        skew = data[i].skew()
        kurt = data[i].kurtosis()
        
        #plt.title('mean= {}'.format(round(mean,2))
        
        plt.title('mean = {}, median = {}\nmean= {},range={}'.format(round(median,2),round(mean,2),round(mean,2),round(ran,2)))
         
        plt.title('min = {}, max = {}\nmean= {},median={},range={}'.format(round(mini,2),round(maxi,2),round(mean,2),
                                                                           round(median,2),round(ran,2)))
         
```

```python
UVA_numeric(data)
```
